```python
def demonstrate_row_major():
    matrix = [
        [8, 9, 10],
        [11, 12, 13],
        [14, 15, 16]
    ]

    rows = 3
    cols = 3
    flat_memory = []

    print("Logical 2D matrix view:")
    for row in matrix:
        print(row)

  
    for i in range(rows):
        for j in range(cols):
            flat_memory.append(matrix[i][j])

    print("\nRow-major memory representation:")
    print(flat_memory)
    print("\nMapping (row-major):")
    index = 0
    for i in range(rows):
        for j in range(cols):
            print(f"matrix[{i}][{j}] -> memory[{index}] = {matrix[i][j]}")
            index += 1



demonstrate_row_major()
```

```python
def demonstrate_column_major():
    matrix = [
        [8, 9, 10],
        [11, 12, 13],
        [14, 15, 16]
    ]

    rows = 3
    cols = 3
    flat_memory = []

    print("Logical 2D matrix view:")
    for row in matrix:
        print(row)


    for j in range(cols):
        for i in range(rows):
            flat_memory.append(matrix[i][j])

    print("\nColumn-major memory representation:")
    print(flat_memory)

   
    print("\nMapping (column-major):")
    index = 0
    for j in range(cols):
        for i in range(rows):
            print(f"matrix[{i}][{j}] -> memory[{index}] = {matrix[i][j]}")
            index += 1


# Call function
demonstrate_column_major()
```
