```python
# Create a list
my_list = []

# Number of elements
n = int(input("Enter the number of elements: "))

# Input list elements
print("Enter the elements:")
for i in range(n):
    element = int(input())
    my_list.append(element)

print("Original List:", my_list)

# Insert an element
position = int(input("Enter the position to insert (0-based index): "))
value = int(input("Enter the element to insert: "))

if position >= 0 and position <= len(my_list):
    my_list.insert(position, value)
    print("List after insertion:", my_list)
else:
    print("Invalid position for insertion!")

# Delete an element
position = int(input("Enter the position to delete (0-based index): "))

if position >= 0 and position < len(my_list):
    deleted = my_list.pop(position)
    print("Deleted element:", deleted)
    print("List after deletion:", my_list)
else:
    print("Invalid position for deletion!")

```
